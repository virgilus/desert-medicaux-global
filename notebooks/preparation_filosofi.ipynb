```python
import os

import pandas as pd
```

```python
# Déclaration des constantes
DATA_DIR = '../data/external/'
PROCESSED_DIR = '../data/processed/'

CITIES_FILENAME = 'base-cc-filosofi-2015.xls'
PS_CITIES_FILENAME = 'PS_Communes.csv'
ANALYSIS_FILENAME = 'analyse.csv'
```

```python
filo_df = pd.read_excel(os.path.join(DATA_DIR, CITIES_FILENAME),dtype={'CODEGEO':str}, skiprows=5)
```

```python
ps_df = pd.read_csv(os.path.join(PROCESSED_DIR, PS_CITIES_FILENAME), # Concaténation du répertoire et du nom de fichier
            sep=';',
            dtype={'codecommuneetablissement':str}
                   )
```

```python
analyse_df = filo_df.merge(ps_df, how='outer', indicator=True,
                          left_on='CODGEO', right_on='codecommuneetablissement')
```

```python
analyse_df = (filo_df.merge(ps_df, how='outer', indicator=True,
                           left_on='CODGEO', right_on='codecommuneetablissement')
             .fillna(value={c: 0 for c in ps_df.columns if c != 'codecommuneetablissement'})
             )
```

```python
analyse_df.to_csv(os.path.join(PROCESSED_DIR, ANALYSIS_FILENAME), sep=';', index= False)
```
